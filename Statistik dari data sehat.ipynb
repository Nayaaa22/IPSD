**Statistik Deskriptif dari DataFrame sehat**

```python
sehat.describe()
```

```python
# menghitung mean
sehat.mean()
```

```python
# menghitung standar deviasi
sehat.std(ddof=1)
```

```python
# menghitung varians
sehat.var(ddof=1)
```

```python
# Daftar kolom yang ingin dihitung
sehat2 = ['Age', 'Sleep Duration', 'Quality of Sleep', 'Physical Activity Level', 'Stress Level', 'Heart Rate']
```

Membuat Box Plot Untuk Semua Kolom

```python
for col in sehat2:
    plt.figure(figsize=(8, 6)) 
    sns.boxplot(y=sehat[col])  # Mengubah orientasi menjadi vertikal
    plt.title(f'Box Plot untuk {col}') 
    plt.show()
```

Membuat Scatter Plot

```python
# Menggunakan Histogram
for col in sehat2:
    plt.figure(figsize=(8, 6))
    sns.histplot(sehat[col], bins=10, kde=True, color='skyblue')  # menambahkan KDE untuk melihat distribusi
    plt.title(f'Histogram untuk {col}')
    plt.xlabel(f'Nilai {col}')
    plt.ylabel('Frekuensi')
    plt.show()
```

Membuat Bar Plot untuk semua kolom

```python
# Menghitung rata-rata dari setiap kolom
means = [sehat[col].mean() for col in sehat2]

plt.figure(figsize=(10, 6))
sns.barplot(x=sehat2, y=means)
plt.title('Bar Plot dari Rata-rata Setiap Kolom')
plt.xlabel('Kolom')
plt.ylabel('Rata-rata Nilai')
plt.xticks(rotation=45) 
plt.show()
```

**Hubungan antara Quality of Sleep dan Stress Level**

```python
correlation_quality_stress = sehat['Quality of Sleep'].corr(sehat['Stress Level'])
print(f'Korelasi antara Quality of Sleep dan Stress Level: {correlation_quality_stress:.2f}')

# Visualisasi dengan scatter plot
plt.figure(figsize=(8, 6))
sns.scatterplot(x='Quality of Sleep', y='Stress Level', data=sehat)
plt.title('Quality of Sleep vs Stress Level')
plt.xlabel('Quality of Sleep')
plt.ylabel('Stress Level')
plt.axhline(y=0, color='red', linestyle='--')  # Garis referensi
plt.axvline(x=0, color='red', linestyle='--')  # Garis referensi
plt.show()
```

Hubungan antara *Quality of Sleep* (Kualitas Tidur) dan *Stress Level* (Tingkat Stres) itu saling memengaruhi. Kualitas tidur yang buruk sering kali membuat seseorang merasa lebih stres, sementara stres yang berlebihan dapat mengganggu tidur yang nyenyak. Akibatnya, kedua faktor ini bisa menciptakan siklus yang merugikan kesehatan mental dan fisik kita. Oleh karena itu, penting untuk menerapkan strategi seperti teknik relaksasi dan rutinitas tidur yang baik agar bisa meningkatkan kualitas tidur dan menurunkan tingkat stres, sehingga kita bisa mencapai kesejahteraan yang lebih baik secara keseluruhan.

**Hubungan antara Sleep Duration dan Heart Rate**

```python
correlation_sleep_heart = sehat['Sleep Duration'].corr(sehat['Heart Rate'])
print(f'Korelasi antara Sleep Duration dan Heart Rate: {correlation_sleep_heart:.2f}')

plt.figure(figsize=(8, 6))
sns.scatterplot(x='Sleep Duration', y='Heart Rate', data=sehat)
plt.title('Sleep Duration vs Heart Rate')
plt.xlabel('Sleep Duration')
plt.ylabel('Heart Rate')
plt.axhline(y=0, color='red', linestyle='--')  # Garis referensi
plt.axvline(x=0, color='red', linestyle='--')  # Garis referensi
plt.show()
```

Hubungan antara *Sleep Duration* (Durasi Tidur) dan *Heart Rate* (Detak Jantung) juga menunjukkan keterkaitan yang menarik. Durasi tidur yang cukup cenderung berhubungan dengan detak jantung yang lebih rendah dan lebih stabil, karena saat tidur, tubuh memiliki kesempatan untuk beristirahat dan memulihkan diri. Sebaliknya, kurang tidur bisa meningkatkan detak jantung, yang bisa jadi tanda stres atau tekanan pada tubuh. Oleh karena itu, penting untuk menjaga durasi tidur yang memadai agar bisa mendukung kesehatan jantung dan kesejahteraan secara keseluruhan.
